```python
#    2. Most expensive trip per mile (total amout/mile)

q = """
        SELECT MAX(Total_amount/Trip_distance) as 'Max Per Mile' 
        FROM nyc_data_sep
        
    """

pysqldf(q)
```

```python
#    3. Most generous trip (highest tip).

q = """
        SELECT MAX(Tip_amount) as 'Max Tip' 
        FROM nyc_data_sep
        
    """

pysqldf(q)
```

```python
#    4. Longest trip duration.

q = """
        SELECT MAX(strftime('%s',lpep_dropoff_datetime) - strftime('%s',Lpep_pickup_datetime)) as 'Longest Trip'
        FROM nyc_data_sep
        
    """

pysqldf(q)

```

```python

#   5. Mean tip by hour

q = """
        
        SELECT strftime('%H', lpep_dropoff_datetime) as trip_hour, 
        AVG(Tip_amount) as avg_tip
        FROM nyc_data_sep
        GROUP BY trip_hour

    """

pysqldf(q).head(2)

```

```python

#   6. Median trip cost (This question is optional. You can search for “median” calculation if you want).


q = """

        SELECT ((MIN(Total_amount) + MAX(Total_amount)) - (MIN(Total_amount))/2) as Median_amount
        FROM (
        SELECT Total_amount
          FROM nyc_data_sep
          ORDER BY Total_amount
          LIMIT 2 - ((SELECT COUNT(*) FROM nyc_data_sep) % 2) -- odd 1, even 2
          OFFSET (SELECT (COUNT(*) - 1) / 2 FROM nyc_data_sep)
);



    """

pysqldf(q)



```

```python
#   7. Average total trip by day of week (Fortunately, we have day of week information. Otherwise, we need to create a new date column without hours from date column. Then, we need to create "day of week" column, i.e Monday, Tuesday .. or 1, 2 ..,  from that new date column. Total trip count should be found for each day, lastly average total trip should be calculated for each day).
 
q = """      
        SELECT COUNT(*) as total_trip_by_day
          FROM nyc_data_sep
          GROUP BY lpep_pickup_day_of_week
          
;
    """

pysqldf(q).head(2)    
    
    
```

```python
#11. Which airport welcomes more passengers: JFK or EWR? Tip: check RateCodeID from data dictionary for the definition (2: JFK, 3: Newark).
 
q = """      
       SELECT 
       CASE
          CASE 
          WHEN RateCodeID = 2  THEN  'JFK'
          ELSE   'EWR'
          END Airport
          
        FROM QW
        WHERE QW=
        
          (
          SELECT SUM(Passenger_count) as total_pass, RateCodeID
          FROM nyc_data_sep
          WHERE RateCodeID==2 or RateCodeID==3 
          GROUP BY RateCodeID
          ORDER BY total_pass DESC
          LIMIT 1
          )
       
;
    """
pysqldf(q).head(2)

```
