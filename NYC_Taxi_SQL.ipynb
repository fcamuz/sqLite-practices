```python
#15. Find the trips of which trip distance is greater than 15 miles (included) or less than 0.1 mile (included). 
#It is possible to write this with only one where statement. However, this time write two queries and "union" them. 
#The purpose of this question is to use union function. You can consider this question as finding outliers in a quick and 
#dirty way, which you would do in your professional life too often.

    
q = """      
        SELECT  Trip_distance
          FROM nyc_data_sep
          WHERE Trip_distance <=0.1   
          
        UNION
        
         SELECT  Trip_distance
          FROM nyc_data_sep
          WHERE Trip_distance >=15 
;
    """
pysqldf(q).tail() 
```

```python
# 16. We would like to see the distribution (not like histogram) of Total_amount. 
#Could you create buckets, or price range, for Total_amount and find how many trips there are in each buckets? 
#Each range would be 5, until 35, i.e. 0-5, 5-10, 10-15 … 30-35, +35. The expected output would be as follows:

q = """      
    select

        bucket_floor,
        (bucket_floor || 'to ' || bucket_ceiling) as payment_range_in_dolars,
        count(*) as Trip_count
    from ( 
        select 
            round(Total_amount/5.00)*5 as bucket_floor,
            round(Total_amount/5.00)*5 + 5 as bucket_ceiling
        from nyc_data_sep
        ) 
       
    group by payment_range_in_dolars
    order by bucket_floor
;
    """ 
pysqldf(q).head()   
```

```python
q = """
        SELECT driver_id, sum(Total_amount) Total_amount, Payment_Range_in_Dollars FROM
        (SELECT Total_amount, driver_id,
               CASE WHEN Total_amount < 5 THEN '0 - 5'
                    WHEN Total_amount >= 5 AND Total_amount < 10 THEN '10 - 15'
                    WHEN Total_amount >= 10 AND Total_amount < 15 THEN '10 - 15'
                    WHEN Total_amount >= 15 AND Total_amount < 20 THEN '15 - 20'
                    WHEN Total_amount >= 20 AND Total_amount < 25 THEN '20 - 25'
                    WHEN Total_amount >= 25 AND Total_amount < 30 THEN '25 - 30'
                    WHEN Total_amount >= 30 AND Total_amount < 35 THEN '30 - 35'
                    WHEN Total_amount >= 35 THEN '35+' END as Payment_Range_in_Dollars
        FROM nyc_data_sep)
        GROUP BY Payment_Range_in_Dollars
        
    """
pysqldf(q)
```

```python
#18. Could you find the highest 3 Total_amount trips for each driver? Hint: Use “Window” functions.

   
q = """      
     SELECT rs.Total_amount, rs.driver_id
      from ( 
        select 
            driver_id, Total_amount,
            Rank() over (Partition BY driver_id
                ORDER BY Total_amount DESC ) AS Rank
        from nyc_data_sep   
        )  rs where Rank <=3

    
    """

pysqldf(q)
```

```python
#18. Could you find the lowest 3 Total_amount trips for each driver? Hint: Use “Window” functions.

   
q = """      
     SELECT rs.Total_amount, rs.driver_id, Rank
    
      from ( 
        select 
            driver_id, Total_amount,
            dense_rank() over (Partition BY driver_id
                ORDER BY Total_amount ASC ) AS Rank
        from nyc_data_sep   
        )  rs where Rank <=3

    
    """

pysqldf(q)
```

```python
# 20. Could you find the lowest 10 Total_amount trips for driver_id 1? 
# +Do you see any anomaly in the rank? (same rank, missing rank etc). 
# Could you “fix” that so that ranking would be 1, 2, 3, 4… 
# (without any missing rank)? Note that 1 is the lowest Total_amount in this question.
#Also, same ranks would continue to exist since there might be the 
#same Total_amount. Hint: dense_rank.


     
q = """      
     SELECT DISTINCT rs.Total_amount, rs.driver_id, Rank
    
      from ( 
        select 
            driver_id, Total_amount,
            DENSE_RANK() over (Partition BY driver_id
                ORDER BY Total_amount asc) AS Rank
        from nyc_data_sep   
        )  rs where  driver_id=1
        LIMIT 10
            
    
    """

pysqldf(q)
```

```python
# 26. Revenue of drivers each month.
    
q = """  

    SELECT   coalesce(oct, 0), coalesce(sep, 0), coalesce( oct-sep, 0) , o.driver_id
    
    From 
    (
    
    SELECT sum(Total_amount) oct , driver_id
    FROM 
    nyc_data_oct
    group by driver_id
    ) o
    
     left OUTER JOIN
    
    (
    SELECT sum(Total_amount) sep, driver_id
    FROM 
    nyc_data_sep
    group by driver_id) s
    on  o.driver_id=s.driver_id
    
    


    """

pysqldf(q) 
```

```python
# 29. Revenue per day of week comparison.

    
q = """  

    SELECT  coalesce(oct, 0), coalesce(sep, 0),coalesce( oct-sep, 0), o.lpep_pickup_day_of_week day_of_week
    
    From 
    (
    
    SELECT sum(Total_amount) oct , driver_id,lpep_pickup_day_of_week
    FROM 
    nyc_data_oct
    group by lpep_pickup_day_of_week
    ) o
    
    left OUTER JOIN
    
    (
    SELECT sum(Total_amount) sep, driver_id,lpep_pickup_day_of_week
    FROM 
    nyc_data_sep
    group by lpep_pickup_day_of_week) s
   
    on  o.lpep_pickup_day_of_week=s.lpep_pickup_day_of_week


    """

pysqldf(q)     
```
