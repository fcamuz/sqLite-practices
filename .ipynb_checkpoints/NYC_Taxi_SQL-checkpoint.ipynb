```python
#11. Which airport welcomes more passengers: JFK or EWR? Tip: check RateCodeID from data dictionary for the definition (2: JFK, 3: Newark).
 
q = """      
       SELECT 
          RateCodeID, 
          CASE RateCodeID
          WHEN RateCodeID =2 THEN  'JFK'
          ELSE   'EWR'
          END Airport

          
        FROM 
          (
          SELECT SUM(Passenger_count) as total_pass, RateCodeID
          FROM nyc_data_sep
          WHERE RateCodeID==2 or RateCodeID==3 
          GROUP BY RateCodeID
          ORDER BY total_pass DESC
          LIMIT 1
          )
       
;
    """
pysqldf(q).head(2)

```

```python
#15. Find the trips of which trip distance is greater than 15 miles (included) or less than 0.1 mile (included). 
#It is possible to write this with only one where statement. However, this time write two queries and "union" them. 
#The purpose of this question is to use union function. You can consider this question as finding outliers in a quick and 
#dirty way, which you would do in your professional life too often.

    
q = """      
        SELECT  Trip_distance
          FROM nyc_data_sep
          WHERE Trip_distance <=0.1   
          
        UNION
        
         SELECT  Trip_distance
          FROM nyc_data_sep
          WHERE Trip_distance >=15 
;
    """
pysqldf(q).tail() 
```

```python
# 16. We would like to see the distribution (not like histogram) of Total_amount. Could you create buckets, or price range, for Total_amount and find how many trips there are in each buckets? Each range would be 5, until 35, i.e. 0-5, 5-10, 10-15 … 30-35, +35. The expected output would be as follows:

q = """      
    select

        bucket_floor,
        (bucket_floor || 'to ' || bucket_ceiling) as payment_range_in_dolars,
        count(*) as Trip_count
    from ( 
        select 
            round(Total_amount/5.00)*5 as bucket_floor,
            round(Total_amount/5.00)*5 + 5 as bucket_ceiling
        from nyc_data_sep
        ) 
       
    group by payment_range_in_dolars
    order by bucket_floor
;
    """ 
pysqldf(q).head()   
```

```python
#17. We also would like to analyze the performance of each driver’s earning. 
#Could you add driver_id to payment distribution table?  The expected output would be as follows:


q = """      
    select
        driver_id,
        (bucket_floor || 'to ' || bucket_ceiling) as payment_range_in_dolars, 
        count(*) as Trip_count
    from ( 
        select 
            driver_id, 
            round(Total_amount/5.00)*5 as bucket_floor,
            round(Total_amount/5.00)*5 + 5 as bucket_ceiling
        from nyc_data_sep
        ) 
    
    group by payment_range_in_dolars
    order by driver_id, bucket_floor
;
    """ 
pysqldf(q)
        
        
        
```
